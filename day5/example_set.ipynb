```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filepath="/content/titanic.csv"
```

```python
df=pd.read_csv(filepath)
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().value_counts()
```

```python
df.shape
```

```python
missing=df.isnull().sum()
print(missing)
```

```python
ax=missing.plot(kind='bar',color='blue')
plt.title('Missing values per coloumn')
plt.xlabel('column')
plt.ylabel('Missing values')
plt.xticks(rotation=45)

for index,value in enumerate(missing):
  ax.text(index,value,str(value),ha='center',va='bottom',fontsize=10,fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
df=df.drop(columns=['Cabin'])
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
survivalc=df['Survived'].value_counts()
print("Survived\n",survivalc)
```

```python
sns.countplot(x='Survived', data=df, palette='pastel')
plt.title('Survival Count (0 = Died, 1 = Survived)')
plt.xlabel('Survived')
plt.ylabel('Passenger Count')
plt.show()
```

```python
genderc=df['Sex'].value_counts()
print("Gender count:\n",genderc)
```

```python
survivalbyg=df.groupby(['Sex','Survived']).size()
print("Survival Count By Gender:\n",genderc)
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Passenger Count')
plt.legend(title='Survived')
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts()
print("\nPassenger Class Counts:\n", pclass_counts)
```

```python
labels = ['Class 1 (Upper)', 'Class 2 (Middle)', 'Class 3 (Lower)']

# Plotting the pie chart
plt.figure(figsize=(7, 7))
plt.pie(pclass_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'seashell', 'olive'])
plt.title('Passenger Class Distribution (Pclass)', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
sns.kdeplot(df['Age'], fill=True, color='skyblue')
plt.title('Age Distribution (KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=20, palette='Set1')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 40, 60, 100], labels=['Child', 'Teen', 'Adult', 'Middle Age', 'Senior'])
sns.countplot(x='AgeGroup', hue='Survived', data=df)
plt.title('Survival by Age Group')
plt.show()
```
